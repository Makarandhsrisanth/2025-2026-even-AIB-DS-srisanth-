```python

numbers = {1, 2, 3, 4}


print(numbers)


numbers.add(5)


numbers.remove(2)


if 3 in numbers:
    print("3 is in the set")


for num in numbers:
    print(num)
```

```python

student = {
    "name": "Alice",
    "age": 20,
    "course": "Python"
}


print(student)


print(student["name"])

student["grade"] = "A"


student["age"] = 21


for key, value in student.items():
    print(key, ":", value)
```
