```python
import pandas as pd

# data manipulation operations
data = {
    "NAME": ["A", "B", "C"],
    "AGE": [20, 21, 23],
    "MARKS": [70, 80, 90]
}

df = pd.DataFrame(data)
print("original data:")
print(df)

# adding new column
df["BONUS"] = 5
print("\nafter adding bonus:")
print(df)

# increase marks
df["MARKS"] = df["MARKS"] + 2
print("\nafter increasing marks:")
print(df)

# create new column from existing columns
df["TOTAL"] = df["MARKS"] + df["BONUS"]
print("\nafter adding total:")
print(df)

# change value
df.loc[0, "AGE"] = 21
print("\nafter changing age:")
print(df)

# rename column
df.rename(columns={"TOTAL": "FINAL"}, inplace=True)
print("\nafter renaming column:")
print(df)

# remove column
df.drop("BONUS", axis=1, inplace=True)
print("\nafter removing bonus:")
print(df)
```

```python
# Selection Sort

n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input("Enter element: ")))

print("Original array:", arr)

for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted array:", arr)
```

```python
# Bubble Sort

n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input("Enter element: ")))

print("Original array:", arr)

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print("Sorted array:", arr)
```

```python
# Linear Search

n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

key = int(input("Enter element to search: "))

found = False

for i in range(n):
    if arr[i] == key:
        print(f"Element {key} found at position {i+1}")
        found = True
        break

if not found:
    print("Element not found")
```

```python
# Binary Search

n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

arr.sort()
print("Sorted list:", arr)

key = int(input("Enter the key: "))

low = 0
high = n - 1
found = False

while low <= high:
    mid = (low + high) // 2

    if arr[mid] == key:
        print(f"Element {key} found at position {mid + 1}")
        found = True
        break
    elif key < arr[mid]:
        high = mid - 1
    else:
        low = mid + 1

if not found:
    print("Element not found")
```

```python
import pandas as pd

# Grouping -> sum, mean, count
data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [80, 85, 75, 90, 88]
}

df = pd.DataFrame(data)
print("The original table:")
print(df)

# Avg marks department wise (mean)
print("\nAverage marks by department:")
print(df.groupby("Department")["Marks"].mean())

# Total marks by department (sum)
print("\nTotal marks by department:")
print(df.groupby("Department")["Marks"].sum())

# Count of students by department
print("\nCount by department:")
print(df.groupby("Department")["Marks"].count())
```
